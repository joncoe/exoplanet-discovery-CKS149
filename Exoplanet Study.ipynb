```python
import pandas as pd
import numpy as np
# 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
# 
import chart_studio as py
import plotly.graph_objs as go 
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
dataFile = "exoplanets-data-23-12-CONFIRMED.csv"
exoplanets = pd.DataFrame(pd.read_csv(dataFile, low_memory=False))
exoplanets
```

```python
exoplanets.describe()
```

```python
unique_planets_df = exoplanets.groupby('pl_name').first().reset_index()
unique_planets_df.describe()
```

```python
len(unique_planets_df.index)
```

```python
correlations = unique_planets_df.corr(numeric_only=True)
correlations
```

```python
filter_cols = ['pl_name','gaia_id', 'sy_snum', 'sy_pnum', 'sy_mnum', 'cb_flag', 'pl_orbper', 'pl_orbperlim', 'sy_dist', 'sy_gaiamag', 'pl_rade', 'pl_radj', 'discoverymethod', 'sy_gaiamag', 'pl_trandur']
filtered_exos = unique_planets_df.filter(filter_cols, axis=1)
filtered_exos['pl_orbper'] = filtered_exos['pl_orbper'].astype(float)
filtered_exos['pl_rade'] = filtered_exos['pl_rade'].astype(float)
filtered_exos.sample(15)
```

```python
sns.heatmap(filtered_exos.corr(),cmap='coolwarm')
```
