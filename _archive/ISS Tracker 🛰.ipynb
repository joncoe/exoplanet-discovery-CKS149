```python
import pandas as pd
# all active satellites 
active_satellites = "https://celestrak.org/NORAD/elements/gp.php?GROUP=active&FORMAT=json"
active_df = pd.read_json(active_satellites)
active_df.sample(10)
```

## Let's filter all active satellites for starlink

```python
starlink = active_df[active_df['OBJECT_NAME'].str.contains('STARLINK')]
starlink.sample(10)
```

```python
total_starlink = starlink.shape[0]
print("Right now Nov 16 2023 there are " + str(total_starlink) + " Starlinks in orbit")
```
